```python
import numpy as np

# -------------------------
# Sigmoid & Derivative
# -------------------------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(output):
    return output * (1 - output)

# -------------------------
# Inputs
# -------------------------
x1 = 0.35
x2 = 0.9
inputs = np.array([x1, x2])

# -------------------------
# Weights (from diagram)
# -------------------------
# Input → Hidden
w13 = 0.1
w23 = 0.8
w14 = 0.4
w24 = 0.6

# Hidden → Output
w35 = 0.3
w45 = 0.9

# -------------------------
# Target & Learning Rate
# -------------------------
target = 0.5
lr = 1

# =====================================================
# FORWARD PASS
# =====================================================
# Hidden layer
net_h3 = x1 * w13 + x2 * w23
net_h4 = x1 * w14 + x2 * w24

h3 = sigmoid(net_h3)
h4 = sigmoid(net_h4)

# Output layer
net_o5 = h3 * w35 + h4 * w45
output = sigmoid(net_o5)

print("Forward Pass:")
print("H3:", h3)
print("H4:", h4)
print("Output:", output)

# =====================================================
# BACKWARD PASS
# =====================================================
# Output layer error
error = target - output
delta_o5 = error * sigmoid_derivative(output)

# Hidden layer errors
delta_h3 = sigmoid_derivative(h3) * (delta_o5 * w35)
delta_h4 = sigmoid_derivative(h4) * (delta_o5 * w45)

# =====================================================
# WEIGHT UPDATES
# =====================================================
# Hidden → Output
w35 += lr * delta_o5 * h3
w45 += lr * delta_o5 * h4

# Input → Hidden
w13 += lr * delta_h3 * x1
w23 += lr * delta_h3 * x2
w14 += lr * delta_h4 * x1
w24 += lr * delta_h4 * x2

print("\nUpdated Weights:")
print("w13:", w13, "w23:", w23)
print("w14:", w14, "w24:", w24)
print("w35:", w35, "w45:", w45)
```
